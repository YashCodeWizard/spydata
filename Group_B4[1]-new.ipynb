```python
 import plotly.express as px
 import pandas as pd
 import plotly.figure_factory as ff
 import numpy as np
 import seaborn as sns
 import matplotlib.pyplot as plt
```

```python
 df = pd.read_csv('heart.csv')
```

```python
df
```

```python
df.isna().sum()
```

**Heatmap**

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),cmap = 'YlGnBu', annot = True)
```

**2)line plot**

```python
 # Max Heart Rate Trend by Age
 plt.figure(figsize=(8,6))
 sns.lineplot(data=df.sort_values(by="age"),
 x="age", y="thalachh", marker="o",
 color="blue")
 plt.title("Max Heart Rate Trend by Age")
 plt.xlabel("Age")
 plt.ylabel("Max Heart Rate")
 plt.show()
```

```python
 # Cholesterol Levels Across Age
 plt.figure(figsize=(8,6))
 sns.lineplot(data=df.sort_values(by="age"),
 x="age", y="chol", marker="o",color="red")
 plt.title("Cholesterol Levels Across Age")
 plt.xlabel("Age")
 plt.ylabel("Cholesterol")
 plt.show()
```

```python
 # Resting Blood Pressure Trend by Age
 plt.figure(figsize=(8,6))
 sns.lineplot(data=df.sort_values(by="age"),
 x="age", y="trtbps", marker="o",
 color="green")
 plt.title("Resting Blood Pressure Across Age")
 plt.xlabel("Age")
 plt.ylabel("Resting Blood Pressure")
 plt.show()
```

**3)Bar plot**

```python
 # Average Cholesterol Levels by Chest Pain Type
 plt.figure(figsize=(8,6))
 sns.barplot(data=df, x="cp", y="chol", palette="coolwarm")
 plt.title("Average Cholesterol Levels by Chest Pain Type")
 plt.xlabel("Chest Pain Type")
 plt.ylabel("Cholesterol Level")
 plt.show()
```

**4)Histogram**

```python
 #Age Distribution by Heart Disease (Seaborn)
 plt.figure(figsize=(8,6))
 sns.histplot(df, x="age", hue="output",
 kde=True, bins=20, palette="deep")
 plt.title("Age Distribution by Heart Disease")
 plt.xlabel("Age")
 plt.ylabel("Count")
 plt.show()
```

**5)Scatter Plot**

```python
 # Age vs Max Heart Rate (Seaborn)
 plt.figure(figsize=(8,6))
 sns.scatterplot(data=df, x="age", y="thalachh",
 hue="output", palette="mako")
 plt.title("Age vs Max Heart Rate by Heart Disease")
 plt.xlabel("Age")
 plt.ylabel("Max Heart Rate")
 plt.show()
```

**6)Violin plot**

```python
 # Violin Plot- Cholesterol Level Distribution (Seaborn)
 plt.figure(figsize=(8,6))
 sns.violinplot(data=df, y="chol",
 inner="point", palette="coolwarm")
 plt.title("Cholesterol Level Distribution")
 plt.ylabel("Cholesterol Level")
 plt.show()
```

**7)Strip plot**

```python
 # Strip Plot- Oldpeak Levels by Heart Disease (Seaborn)
 plt.figure(figsize=(8,6))
 sns.stripplot(data=df, x="output", y="oldpeak",
 hue="output", palette="bright", jitter=True)
 plt.title("Oldpeak Levels by Heart Disease")
 plt.xlabel("Heart Disease (Output)")
 plt.ylabel("Oldpeak Level")
 plt.show()
```

**8)Box Plot**

```python
 # Box Plot- Max Heart Rate by Heart Disease (Seaborn)
 plt.figure(figsize=(8,6))
 sns.boxplot(data=df, x="output", y="thalachh",
 palette="bright")
 plt.title("Max Heart Rate by Heart Disease")
 plt.xlabel("Heart Disease (Output)")
 plt.ylabel("Max Heart Rate")
 plt.show()
```

**9)Cluster Plot**

```python
 from sklearn.cluster import KMeans
 # Selecting features for clustering
 X = df[['age', 'thalachh']]
 # Applying K-Means clustering
 kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
 df['cluster'] = kmeans.fit_predict(X)
```

```python
 # Scatter plot with clusters
 plt.figure(figsize=(8,6))
 sns.scatterplot(data=df, x="age", y="thalachh",
 hue="cluster", palette="coolwarm",
 style="cluster", s=100)
 plt.title("K-Means Clustering: Age vs. Max Heart Rate")
 plt.xlabel("Age")
 plt.ylabel("Max Heart Rate")
 plt.legend(title="Cluster")
 plt.show()
```

**10)Swarm Plot**

```python
plt.figure(figsize=(8,6))
sns.swarmplot(data=df, x="output", y="thalachh", palette="Set1")
plt.title("Max Heart Rate by Heart Disease (Swarm Plot)")
plt.xlabel("Heart Disease (Output)")
plt.ylabel("Max Heart Rate")
plt.show()
```
